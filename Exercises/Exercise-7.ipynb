```python
import pandas as pd
# Create sample employee data
employees = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry', 'Iris', 'Jack'],
    'department': ['Sales', 'IT', 'HR', 'Finance', 'Sales', 'IT', 'HR', 'Finance', 'Sales', 'IT'],
    'salary': [45000, 65000, 42000, 58000, 52000, 75000, 48000, 62000, 55000, 72000]
}

df = pd.DataFrame(employees)
print("Employee DataFrame:")
print(df)
```

# Filter employees with salary above the 75th percentile

```python
# Calculate the 75th percentile of salary
percentile_75 = df['salary'].quantile(0.75)
print(f"75th Percentile Salary: ${percentile_75:,.2f}")
```

```python
# Filter employees with salary above the 75th percentile
high_earners = df[df['salary'] > percentile_75]
print(f"\nEmployees with salary above 75th percentile (>${percentile_75:,}):")
print(high_earners)
```
